## 1. Population of each state

```python
import pandas as pd
import numpy as np
import warnings
warnings.filterwarnings('ignore')
```

```python
url="https://en.wikipedia.org/wiki/List_of_states_and_union_territories_of_India_by_population"
df_1=pd.read_html(url)
```

```python
df_1[1]
```

```python
df_1[1].head()
```

Null values are there in National Share (%),Percent rural,Percent Urban columns,so we will remove this columns.

```python
df_1[1].info()
```

```python
df_1[1].isnull().sum()
```

```python
df_1[1].dtypes
```

```python
df_1[1]=df_1[1].dropna(axis=1)
df_1[1] = df_1[1].reset_index(drop=True)
df_1[1]
```

So we have removed null values columns National Share (%),Percent rural,Percent urban columns.
